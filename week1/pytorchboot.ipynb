```python
import torch

x = torch.empty((3,3,3))

print(x)
```

```python
x = torch.rand((3,3))

print(x)
```

```python
x = torch.zeros((3,3))

print(x)
```

```python
x = torch.ones((3,3))

print(x)
print(x.dtype)
print(x.shape)
print(x.device)
```

```python
x = torch.ones((3,3), dtype=torch.int)
print(x)
```

```python
x = torch.ones((3,3))
y = torch.rand((3,3))
z = x + y

print(x)
print(y)
print(f"x+y=\n{z}")
```

```python
x = torch.rand((3,3))
y = torch.rand((3,3))
z = torch.mul(x,y)

print(x)
print(y)
print(f"x*y=\n{z}")
```

```python
x = torch.rand((7,7))

print(x)
print(x[:,0]) # first column
print(x[0,:]) # first row
print(x[:,-1]) # last column
```

```python
x = torch.rand((8,8))
y = x.view((2,32))

print(x)
print(x.shape)
print(y)
print(y.shape)
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0,10,0.01)
y = np.sin(np.pi*x)

plt.plot(x,y)
```
